```python
import os
import glob
import tqdm
import librosa
import torchaudio
import pandas as pd
```

```python
selected_cols = ['READER', 'GENDER', 'AGE', 'ACCENT', 'NUM_SAMPLES', 'TOTAL_SECONDS']
```

```python
def get_duration(audio_path):
    info = torchaudio.info(audio_path)
    duration = info.num_frames / info.sample_rate
    return duration
```

```python
df = pd.read_csv(f'./metadata/speakers.csv')
out_metadata_path = f'./metadata/metadata.csv'
clean_df = []

for _, row in tqdm.tqdm(df.iterrows(), total=len(df)):
    reader = row['READER']
    gender = row['GENDER']
    age = None
    accent = 'Spanish'
    num_samples = len(glob.glob(f'./data/{reader}/**/*.wav', recursive=True))
    total_seconds = [get_duration(wav_path) for wav_path in glob.glob(f'./data/{reader}/**/*.wav', recursive=True)]
    total_seconds = sum(total_seconds)
    clean_df.append({
        'READER': reader,
        'GENDER': gender,
        'AGE': age,
        'ACCENT': accent,
        'NUM_SAMPLES': num_samples,
        'TOTAL_SECONDS': total_seconds
    })
   

clean_df = pd.DataFrame(clean_df)
if 'AGE' in clean_df.columns:
    clean_df['AGE'] = clean_df['AGE'].astype('Int64')
    clean_df['TOTAL_SECONDS'] = clean_df['TOTAL_SECONDS'].astype('float').round(2)  
clean_df.to_csv(f'./metadata/metadata.csv', index=False)
```

```python
print("Total duration of all samples: {:.2f} hours".format(clean_df['TOTAL_SECONDS'].sum() / 3600))
print("Num speakers: {}".format(clean_df['READER'].nunique()))
```
